## Ok, First Let's just build the network from scratch

```python
# defining some hyperparameters
BATCH_SIZE = 256
config = [2, 2, 4, 1]
ACTIVATION : Literal['sigmoid', 'relu'] = 'sigmoid'
OUT_ACTICATION = 'sigmoid'
```

```python
def initialize_hidden_parameters(
        config: list[int],
):
    weights = []
    baises = []
    for i in range(len(config)-1):
        w = np.random.randint(size=(config[i], config[i+1]), low=0, high=10)
        b = np.random.randint(size=(1, config[i+1]), low=0, high=10)

        weights.append(w)
        baises.append(b)

    return weights, baises
```

```python
weights, baises = initialize_hidden_parameters(config=config)
```

```python
def sigmoid(Z:np.array):
    return 1/(1+np.exp(-Z))
```

```python
def forward_propagation(
        X: np.array,
		weights: np.array,
        baises: np.array,
):
    activations = X
    print(f"Data: {X}\n")
    for i in range(len(weights)):
        Z = activations @ weights[i] + baises[i]
        print(f"Layer {i+1}: Z: {Z}\n")
        activations = sigmoid(Z)
        print(f"Layer {i+1}: activations: {activations}\n")
        
    return activations
```

```python
weights, baises = initialize_hidden_parameters(config=config)
```

```python
X = np.random.randint(size=(3, 2), low=0, high=5)
```

```python
output = forward_propagation(
    X=X,
    weights=weights,
    baises=baises
)
```

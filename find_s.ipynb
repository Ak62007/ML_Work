```python
import pandas as pd
import numpy as np
```

```python
data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Cool', 'Change', 'Yes']
]

columns = ['Sky', 'AirTemp', 'Humidity', 'Wind', 'Water', 'Forecast', 'EnjoySport']
df = pd.DataFrame(data, columns=columns)
```

```python
df
```

##### So, we have to start with most specific case

```python
start = ['0']*df.shape[1]
start
```

```python
def find_S(start: list[str], data:pd.DataFrame) -> list[str]:    
    for row in df.iterrows():
        features = row[1][:-1]
        label = row[1][-1]
        if label == "No":
            continue
        else:
            for i in range(len(features)):
                if start[i] == '0':
                    start[i] = features.values[i]
                elif (start[i] != '0') and (start[i] != features.values[i]):
                    start[i] = '?'
                elif (start[i] == features.values[i]):
                    continue
    return start
```

```python
find_S(start=start, data=df)
```
